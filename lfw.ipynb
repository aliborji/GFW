```python
import os
import cv2
import os.path as op
```

```python
source = 'lfw'
dest = 'lfw_resize'
```

```python
os.mkdir(dest)
```

```python
face_res = 100
files = os.listdir(source)
count = 0 #len(os.listdir('real_data/'))

for f in files:
    if f.startswith('.'): continue
    
    files = os.listdir(op.join(source, f))

    # print(len(files))
    for fname in files:
        if fname.startswith('.'): continue
        newf = str(count).zfill(5) + '.jpeg'

        face = cv2.imread(op.join(source,f,fname))
        face = face[50:-50,50:-50] 
        face = cv2.resize(face,(face_res,face_res))
        cv2.imwrite(op.join(dest, newf),face)


        count += 1


```
